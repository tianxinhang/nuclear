```python
part_file_list = ['67#', '83#', '61#', '73#']
filter_prefix = '_FFT'
part_name = '67#' + filter_prefix
```

```python
def get_part_data(part_name, key_col=25, drop_col=['set']):
    u_data = pd.DataFrame()
    for csv in glob.glob(part_name+'*'):
        print(csv)
        temp_data = pd.read_csv(csv, index_col=0)
        new_cols = list([c.split('.')[-1].replace('WPS-', '') for c in temp_data.columns[:key_col]])
#         print(new_cols)
        new_cols += list(temp_data.columns[key_col:])
        temp_data.columns = new_cols
        temp_data.drop(drop_col, axis=1, inplace=True)
        
        print(temp_data.columns[:(key_col+5)])
        u_data = pd.concat([u_data, temp_data], axis=0)
        
    u_data['part'] = part_name[:2]
    return u_data

p_data = get_part_data(part_name)
p_data
```
